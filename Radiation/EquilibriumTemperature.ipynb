# Notebook Calcolo della Temperatura di Equilibrio di un corpo

## Calcolo semplificato nel caso di un rapidly rotating body
### Nell'approssimazione iniziale trascura variazioni longitudinali/latitudinali

Assunzione di orbita circolare

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Ab = 0.0
e = 1.0
Lsun = 3.828e26  # W
au2m = 1.496e11 # m/au
sigma = 5.6704e-8 # W/(m^2K^4)

Lsunau = Lsun/au2m/au2m
```

```python
rau = np.linspace(0.2,50.,1000)
teq = np.power((Lsunau/(4.*np.pi*rau*rau)*(1.-Ab)/(4.*e*sigma)),0.25)
```

```python
fig,ax = plt.subplots(2,1,figsize=(10,5))

nrin = np.where(rau<5.5)
nrou = np.where(rau>4.5)

ax[0].plot(rau[nrin],teq[nrin])
ax[0].set_ylim(70.,800.)
#ax[0].set_xlabel(r'Distance (au)')
ax[0].set_ylabel(r'T$_{eq}$ (K)')

ax[1].plot(rau[nrou],teq[nrou])
ax[1].set_ylim(20.,200.)
ax[1].set_xlabel(r'Distance (au)')
ax[1].set_ylabel(r'T$_{eq}$ (K)')

labels = ['Venus','Venus$_{sur}$','Earth','Mars','Jupiter','Saturn','Uranus','Neptune','Pluto']
distance = [0.72, 0.72, 1.0, 1.5, 5.2, 9.5, 19.2, 30.0, 39.5]
temperature = [327, 737, 290., 210., 165., 134., 76., 72., 50.]
for i in [0,1,2,3,4]:
    ax[0].text(distance[i],temperature[i],labels[i],ha='center',va='center')
for i in [4,5,6,7,8]:
    ax[1].text(distance[i],temperature[i],labels[i],ha='center',va='center')
```
